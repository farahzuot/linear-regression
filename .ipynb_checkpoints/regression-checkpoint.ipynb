# linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('test.csv', sep=',')
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
#df['sex'] = np.where(df['smoker'] == 'yes', 1, 0)
#df.head(5)
```

```python
X = df['battery_power'].values.reshape(-1,1)
X
```

```python
Y = df['talk_time'].values
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=100)
```

```python
%matplotlib inline
plt.scatter(x_train,y_train,color='red')
plt.xlabel('speed')
plt.ylabel('ram')
plt.title('data')
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
print(f'Testing Accuracy: {model.score(x_test, y_test)*100}%')
```

```python
print(f'Training Accuracy: {model.score(x_train, y_train)*100}%')
```

```python

%matplotlib inline
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_test, color='y')
plt.plot(x_test, y_predict, color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Training Data')
plt.show()
```
